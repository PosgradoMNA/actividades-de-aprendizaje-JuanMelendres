```python
import numpy as np
import pandas as pd
```

```python
x = { 'Company': ['Ford', 'Ford', 'VW', 'BMW', 'Cooper', 'Cooper'],
     'Stars': [1, 2, np.nan, 2, 1, 1],
     'Weight': [2, 4, 2, 2, 3, None],
     'Origin': ['Mexico', 'China', 'Mexico', None, 'China', np.nan],
     'Length': [50, 40, 30, np.nan, 45, pd.NaT]
    }
```

```python
df2 = pd.DataFrame(data = x)
df2
```

```python
df2.isnull().values.any() # Si tiene datos perdidos
```

```python
df2.isnull().any() # Que columnas tienen datos perdidos
```

```python
ndf = df2.dropna() # Limpiar datos, pero se pierde la informacion de los demas
ndf
```

```python
df2
```

Solucion 1:
Eliminar todos los renglones, si tiene al menos un dato perdido

```python
ndf2 = df2.copy()
ndf2.dropna(inplace = True)
ndf2
# Al final solo me quedan los datos que si estan completos
```

Si existe al menos un dato perdido en una columna, voy a eliminar esta columna

```python
ndf3 = df2.copy()
ndf3.dropna(inplace = True, 
            axis = 1 # aplicar para todas las columnas
            )
ndf3
```

Eliminar renglones si todos los valores estan perdidos

```python
ndf4 = df2.copy()
ndf4.dropna(how = 'all',
            inplace = True)
ndf4
```

si existen al menos 3 datos, no voy a eliminar estos renglones

```python
ndf5 = df2.copy()
ndf5.dropna(thresh = 3,
            inplace = True)
ndf5
```

```python
ndf6 = df2.copy()
ndf6.dropna(thresh = 3,
            inplace = True,
            axis = 1 ) # columna
ndf6
```

Remplazar datos con promedio

```python
ndf7 = df2.copy()
ndf7['Stars'].fillna(value = ndf7['Stars'].mean(), inplace = True)
ndf7
```

Mediana

```python
ndf8 = df2.copy()
sm = ndf8['Stars'].median()
ndf8['Stars'].fillna(value = sm, inplace = True)
ndf8
```
