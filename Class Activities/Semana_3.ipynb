**1. Fundamentos de bases de datos y para ciencia de datos:**
Es muy importante el conocer que es una base de datos ya de estas son de donde sacamos los datos para ser analizadas, por lo que una base de datos es una colección de datos. Dentro de la ciencia de datos se requiere el analizar los datos, la preparación de los datos para su análisis y el poder visualizarlos. También se crean modelos con los datos. Un data warehouse es un alancen central de información el cual se puede analizar para la toma de decisiones.  Un data warehouse solo almacena datos que han sido modelados y estructurados. Solucion 2:

```python
sol_2_df = df.copy() # Creamos una copia del DataFrame
wm = sol_2_df.X2.mean() # Obtenemos la media de la coumna X2
wm
```

```python
sol_2_df['X2'].fillna(value = wm, inplace = True) # Vamos a llenar los datos faltantes con la media
sol_2_df
```

Hay dos grandes porblemas con la media si la utilizamos como remplazo de valores:

*   Problema 1: La imputación de medias no conserva las relaciones entre variables. Entonces, si los datos faltan completamente al azar, la estimación de la media permanece sin sesgo.

*   Problema 2: la imputación de la media conduce a una subestimación de los errores estándar. Cualquier estadística que utilice los datos imputados tendrá un error estándar demasiado bajo.

```python
sol_2_df['X3'].fillna(value = sol_2_df.X3.median(), inplace = True) # Vamos a llenar los datos faltantes con la mediana
sol_2_df
```

```python
mm  = sol_2_df.X4.mode() # Ahora visualizaremos la moda de la columna X4
mm
```

```python
sol_2_df['X4'].fillna(value = mm[0], inplace = True) # Vamos a llenar los datos faltantes con la moda
sol_2_df
```
